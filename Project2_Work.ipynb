```python
class_order = {'Fr': 1, 'So': 2, 'Jr': 3, 'Sr': 4}

# For each dataframe, add a numeric column for sorting
adv_ncaa_stats['cls_numeric'] = adv_ncaa_stats['cls'].map(class_order)
ncaa_stats['cls_numeric'] = ncaa_stats['cls'].map(class_order)

# Keep only the latest class for each player in both dataframes
adv_stats_latest = adv_ncaa_stats.loc[adv_ncaa_stats.groupby('player')['cls_numeric'].idxmax()]
ncaa_stats_latest = ncaa_stats.loc[ncaa_stats.groupby('player')['cls_numeric'].idxmax()]

# Drop the helper column
final_adv_stats = adv_stats_latest.drop('cls_numeric', axis=1)
final_ncaa_stats = ncaa_stats_latest.drop('cls_numeric', axis=1)
```
