# Random Forest

Random forest is an ensemble learning method used for classification that builds multiple different decision trees and classifies based on the average result from each tree's predictions. Each tree is trained on a random sample of the data and uses a random subset of features to predict the classification results. The final prediction is whichever result won the majority of the time when you are solving a classification problem. Some pros are that it handles non-linear relationships well, it reduces overfitting better than a single decision tree, and it works well with larger sets of data. Some cons of this model are that it is typically slower to train because it works with larger datasets, and it is more difficult to understand and interpret than a single decision tree.

```python
# Fit a random forest classifier to the training data
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)
```

```python
# After fitting the model on the training data, begin the classification process with the random forest model

y_pred = rf_model.predict(X_test)
```

```python
# Find the accuracy and details of the model
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
print("ROC-AUC:", roc_auc_score(y_test, rf_model.predict_proba(X_test)[:,1]))
```

```python
# Create feature importances to see the impact each variable has and if any should be removed to potentially increase validity and accuracy of the model
importances = rf_model.feature_importances_

# put them in a df and plot for convenience 
feat_importances = pd.DataFrame({
    "Feature": X_train.columns,
    "Importance": importances
}).sort_values(by="Importance", ascending=False)

print(feat_importances)


plt.figure(figsize=(10,6))
plt.barh(feat_importances["Feature"], feat_importances["Importance"])
plt.gca().invert_yaxis()  # largest on top
plt.title("Random Forest Feature Importances")
plt.show()
```

# XGBoost Model

XGBoost is an advanced boosting algorithim that creates many decision trees, one after another, with each new tree learning from the mistakes of the last, making the last tree the best, theoretically. Each tree learns from the last to become better. It judges past performance by using gradient descent on a loss function and is great at preventing overfitting and it works much faster than some models like Random Forest. Its pros are that it is one of the most advanced models available and can be used for quite complex problems, it handles missing values automatically so they don't skew results, and once again, it is great at dealing with overfitting. Some cons are that it is quite difficult to understand, mostly because of how advanced of a model it is, it also requires precise hyperparameter tuning and training it can be computationally expensive once you begin working with large, complex datasets.

```python
from xgboost import XGBClassifier, plot_importance
```

```python
xg_model = XGBClassifier(random_state = 42)
xg_model.fit(X_train, y_train)
```

```python
xg_y_pred = xg_model.predict(X_test)
```

```python
xg_cm = confusion_matrix(y_test, xg_y_pred)

print("Confusion Matrix: ")
print(xg_cm)
print(classification_report(y_test, xg_y_pred))
print("RPC-AUC:", roc_auc_score(y_test, xg_model.predict_proba(X_test)[:, 1]))
```

```python
# Create feature importances to see the impact each variable has and if any should be removed to potentially increase validity and accuracy of the model
importances = xg_model.feature_importances_

# put them in a df and plot for convenience 
feat_importances = pd.DataFrame({
    "Feature": X_train.columns,
    "Importance": importances
}).sort_values(by="Importance", ascending=False)

print(feat_importances)

# Use gain for the importance type because it shows the average improvement a variable gives the model output
plot_importance(xg_model, importance_type = 'gain', max_num_features = 25)
plt.show
```

# Logistic Regression

Logistic Regression is a lineaar model used for binary classification where you want to sort datapoints into one of two categories. It uses the logistic function to estimate the probability an object is a part of one class and the function's coefficients show how features influence the probability of an outcome. Some of the pros include how easy it is to understand as it is simple to explain how the coeffiencents work and influence the outcome. Additionally it is quick to train and works well with smaller datasets that more advanced models may struggle with. For simple classification problems, it is a top option. Some cons are that it assumes a linear relationship between features and log-odds, it is not good for non-linear or complex relationships, and it does not handle multicollinearity or outliers well.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
```

```python
# To use feature importance with log reg, we need to scale the features so they can be compared on the same scale, ensuring we have the most accurate information

logreg_model = Pipeline([
    ("scaler", StandardScaler()), 
    ("clf", LogisticRegression(max_iter=1000, class_weight="balanced"))
])
logreg_model.fit(X_train, y_train)
log_y_pred = logreg_model.predict(X_test)
lg_cm = confusion_matrix(y_test, log_y_pred)

print(lg_cm)
print(classification_report(y_test, log_y_pred))
print("ROC-AUC:", roc_auc_score(y_test, logreg_model.predict_proba(X_test)[:,1]))
```

```python
# find feature coefficients
coef = logreg_model.named_steps['clf'].coef_[0]

feat_importances = pd.DataFrame({
    "Feature": X_train.columns,
    "Coefficient": coef
}).sort_values(by="Coefficient", ascending=False)
print(feat_importances)

plt.figure(figsize=(10,6))
plt.barh(feat_importances["Feature"], feat_importances["Coefficient"])
plt.gca().invert_yaxis()
plt.title("Logistic Regression Feature Coefficients (Standardized)")
plt.show()
```

# K Nearest Neighbors

KNN is an instance-based algorithim that classifies objects on proximity to training examples. Once given a central point, the algorithm finds the k closest points, or neighbors, in the training data and assigns it the majority class. Its pros are that it's simple to understand and implement in a program, it has no complicated training phase, and it can find complex decision boundaries. Some cons are that it is computationally expensive for large datasets as it spends a lot of power classifying many points, choosing hyperparameters is important and can be difficult to find the right number, and it is heavily influenced by irrelevant features. 

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn_model = KNeighborsClassifier(n_neighbors=5)

knn_model.fit(X_train, y_train)
```

```python
knn_y_pred = knn_model.predict(X_test)
knn_cm = confusion_matrix(y_test, knn_y_pred)

print(knn_cm)
print(classification_report(y_test, knn_y_pred))
print("ROC-AUC:", roc_auc_score(y_test, knn_model.predict_proba(X_test)[:,1]))
```
